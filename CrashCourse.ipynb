### Installing from USB

When your computer begins to boot up, hold down the F12 key to bring up the One-Time Boot Menu.  From here, you can select your USB stick to boot from.  An Ubuntu install screen should appear. Ubuntu is a linux distribution (a.k.a flavor).  

For the most part, you can use the default settings when installing.  If you have more questions, feel free to ask us about details.

Make sure to check the "Install updates" button to make sure your Ubuntu partition is ready to go after installing.  

This will set aside a portion of your hard drive (a partition) to be used for the Ubuntu operating system.  Your Windows operating system installed by default will no longer be able to access this space, so make sure to leave your windows drive some room to expand.  You can drag the sliders around to choose how much space to allocate for Ubuntu -- 50 GB is probably plenty for now.  You can extend the size of your Ubuntu drive later.

The install process shouldn't take more than 30 minutes.  

When you reboot your computer, you'll be presented with a dialog to select which operating system you'd like to use.  If you don't watch as your computer boots up, it will likely choose an operating system for you.  After this installation, Ubuntu will be your default operating system.


### Getting around Linux 

#### Opening up a terminal
The keyboard shortcut ctrl-alt-t should get you to a terminal.  

#### Where am I?
```bash
$ ls
```

#### Getting places
Linux is case sensitive, so make sure to catch your typos.
P.S.: if you start typing a word, and then hit tab, the terminal will help you out with autocomplete.
```bash
$ cd Documents
```

#### Install all the things!
```bash
$ apt-get install git
E: Could not open lock file /var/lib/dpkg/lock - open (13: Permission denied)
E: Unable to lock the administration directory (/var/lib/dpkg/), are you root?
```

In fact, you are not root.  For some operations in linux you need to provide your password in order to make sure you want to complete an action.  Let's try this again!

```bash
$ sudo apt-get install git
```

After you enter in your password, the `git` software package should be installed on your computer.  Think of apt-get like your phone's app store.  Its technically a software repository: a remote location where software installation files are stored.

Let's install a few more things that you'll want to have on your Ubuntu partition:

```bash
$ sudo apt-get python-pip python-numpy python-scipy python-matplotlib ipython ipython-notebook python-pandas python-sympy python-nose
```

Pip is python package manager, much like apt-get is Ubuntu package manager.  Let's install something with pip.  Scikit-learn is a python library (packaged, third party code) for machine learning -- you'll want it for later.

```bash 
$ sudo pip install scikit-learn
```
#### Installing a text editor (your best friend)

To install Sublime, the text editor of choice at Olin, you'll need to add a new repository (source for software) to apt-get.  Then you'll need to run `sudo apt-get update`, which refreshes the list of repositories used by apt-get.  Finally, you can install Sublime with apt-get like you're used to.    

```bash
$ sudo add-apt-repository ppa:webupd8team/sublime-text-3
$ sudo apt-get update
$ sudo apt-get install sublime-text-installer
```

#### Making directories
You can make a new directory (folder) using `mkdir`.
```bash
$ mkdir SHUML # you'll learn the acronym later 
```

#### Make a blank file
And check out Sublime Text!
```bash
$ subl paul 
```

#### Removing files
You probably don't actually want to keep that.
```bash
$ rm paul
```

#### Removing directories
Let's clean up after ourselves.
```bash
$ cd .. # go back a directory
$ rmdir SHUML # remove a file
```
